```python
filepath= r"./output/"
df_budget = pd.read_csv(filepath+'table_budget_ch4.csv',index_col=0)
df_budget['Feedback factor']=None
df_budget
```

```python
df_feedback = pd.read_csv(filepath+'feedback_factor_ch4.csv',index_col=0)
df_feedback
```

```python
df_budget['Feedback factor'].loc['CTRL']=df_feedback['antro01'].loc['OSLOCTM']

scen_list = ['SSP119','SSP434','SSP585']
for scen in scen_list:
    df_budget['Feedback factor'].loc['CNTR_'+scen] = df_feedback[scen].loc['OSLOCTM']
df_budget
```

```python
df_budget['Perturbation lifetime'] = df_budget['Total CH4 lifetime [yrs]']* df_budget['Feedback factor']

#df_budget = df_budget.reindex(sorted_array)
df_budget
```
