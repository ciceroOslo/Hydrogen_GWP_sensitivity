```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '{:,.4f}'.format(x) if abs(x)<0 else ('{:,.2f}'.format(x) if abs(x)<10 else ('{:,.1f}'.format(x) if abs(x)<100 else '{:,.0f}'.format(x))))
```

```python
outputpath= r"./output/"
```

```python
df_lifetime = pd.read_csv(outputpath + 'for_lifetimetable.csv',sep=',',index_col=0)
df_lifetime
```

```python
df_lifetime['h2_loss_delta'] = df_lifetime['h2_loss']-df_lifetime.loc['CNTR']['h2_loss']
df_lifetime['h2_atm_loss_delta'] = df_lifetime['h2_atm_loss']-df_lifetime.loc['CNTR']['h2_atm_loss']
df_lifetime['h2_soil_sink_loss_delta'] = df_lifetime['h2_soil_sink_loss']-df_lifetime.loc['CNTR']['h2_soil_sink_loss']
df_lifetime
```

```python
df_lifetime_ssp = pd.read_csv(outputpath + 'for_lifetimetable_ssp.csv',sep=',',index_col=0)
df_lifetime_ssp
```

```python
ssps = ['SSP119','SSP434','SSP585']
df_lifetime_ssp['h2_loss_delta']=np.nan
df_lifetime_ssp['h2_atm_loss_delta']=np.nan
df_lifetime_ssp['h2_soil_sink_loss_delta']=np.nan


for ssp in ssps:
    name = 'antro10_'+ssp
    df_lifetime_ssp.loc[name,'h2_loss_delta'] = df_lifetime_ssp.loc[name]['h2_loss']-df_lifetime_ssp.loc['CNTR_'+ssp]['h2_loss']
    df_lifetime_ssp.loc[name,'h2_atm_loss_delta'] = df_lifetime_ssp.loc[name]['h2_atm_loss']-df_lifetime_ssp.loc['CNTR_'+ssp]['h2_atm_loss']
    df_lifetime_ssp.loc[name,'h2_soil_sink_loss_delta'] = df_lifetime_ssp.loc[name]['h2_soil_sink_loss']-df_lifetime_ssp.loc['CNTR_'+ssp]['h2_soil_sink_loss']

df_lifetime_ssp
 
```

Lifetime and perturbation lifetime table

```python
columns = ['H2 total lifetime [yrs]', 
           'H2 perturbation lifetime [yrs]', 
           'delta tot [yrs]',
           'delta tot [%]',
           'H2 atm lifetime [yrs]', 
           'H2 atm. perturbation lifetime [yrs]', 
           'delta atm [yrs]',
           'delta atm [%]',
           'H2 drydep lifetime [yrs]', 
           'H2 drydep perturbation lifetime [yrs]',
           'delta drydep [yrs]',
           'delta drydep [%]']

df_index = np.concatenate((df_lifetime.index.values ,ssps))

table_lifetime = pd.DataFrame([],index=df_index,columns=columns)
```

```python
lifetime = df_lifetime.loc['CNTR']['h2_burden']/df_lifetime.loc['CNTR']['h2_loss']
atm_lifetime = df_lifetime.loc['CNTR']['h2_burden']/df_lifetime.loc['CNTR']['h2_atm_loss']
soil_sink_lifetime = df_lifetime.loc['CNTR']['h2_burden']/df_lifetime.loc['CNTR']['h2_soil_sink_loss']

for name in df_lifetime.drop('CNTR').index:
    table_lifetime.loc[name,'H2 total lifetime [yrs]'] = lifetime
    table_lifetime.loc[name,'H2 atm lifetime [yrs]'] = atm_lifetime    
    table_lifetime.loc[name,'H2 drydep lifetime [yrs]'] = soil_sink_lifetime


    table_lifetime.loc[name,'H2 perturbation lifetime [yrs]'] = df_lifetime.loc[name]['h2_burden_delta']/df_lifetime.loc[name]['h2_loss_delta']
    table_lifetime.loc[name,'H2 atm. perturbation lifetime [yrs]'] = df_lifetime.loc[name]['h2_burden_delta']/df_lifetime.loc[name]['h2_atm_loss_delta']
    table_lifetime.loc[name,'H2 drydep perturbation lifetime [yrs]'] =  df_lifetime.loc[name]['h2_burden_delta']/df_lifetime.loc[name]['h2_soil_sink_loss_delta']

for name in ssps:
    table_lifetime.loc[name,'H2 total lifetime [yrs]'] = df_lifetime_ssp.loc['CNTR_'+name]['h2_burden']/df_lifetime_ssp.loc['CNTR_'+name]['h2_loss']
    table_lifetime.loc[name,'H2 atm lifetime [yrs]'] = df_lifetime_ssp.loc['CNTR_'+name]['h2_burden']/df_lifetime_ssp.loc['CNTR_'+name]['h2_atm_loss']  
    table_lifetime.loc[name,'H2 drydep lifetime [yrs]'] = df_lifetime_ssp.loc['CNTR_'+name]['h2_burden']/df_lifetime_ssp.loc['CNTR_'+name]['h2_soil_sink_loss']


    table_lifetime.loc[name,'H2 perturbation lifetime [yrs]'] = df_lifetime_ssp.loc['antro10_'+name]['h2_burden_delta']/df_lifetime_ssp.loc['antro10_'+name]['h2_loss_delta']
    table_lifetime.loc[name,'H2 atm. perturbation lifetime [yrs]'] = df_lifetime_ssp.loc['antro10_'+name]['h2_burden_delta']/df_lifetime_ssp.loc['antro10_'+name]['h2_atm_loss_delta']
    table_lifetime.loc[name,'H2 drydep perturbation lifetime [yrs]'] =  df_lifetime_ssp.loc['antro10_'+name]['h2_burden_delta']/df_lifetime_ssp.loc['antro10_'+name]['h2_soil_sink_loss_delta']


table_lifetime['delta tot [%]']=100.0*(table_lifetime['H2 perturbation lifetime [yrs]']-table_lifetime['H2 total lifetime [yrs]'])/table_lifetime['H2 total lifetime [yrs]']
table_lifetime['delta atm [%]']=100.0*(table_lifetime['H2 atm. perturbation lifetime [yrs]']-table_lifetime['H2 atm lifetime [yrs]'])/table_lifetime['H2 atm lifetime [yrs]']
table_lifetime['delta drydep [%]']=100.0*(table_lifetime['H2 drydep perturbation lifetime [yrs]']-table_lifetime['H2 drydep lifetime [yrs]'])/table_lifetime['H2 drydep lifetime [yrs]']


table_lifetime['delta tot [yrs]'] = table_lifetime['H2 perturbation lifetime [yrs]']  - table_lifetime['H2 total lifetime [yrs]'] 
table_lifetime['delta atm [yrs]'] = table_lifetime['H2 atm. perturbation lifetime [yrs]']  - table_lifetime['H2 atm lifetime [yrs]'] 
table_lifetime['delta drydep [yrs]'] = table_lifetime['H2 drydep perturbation lifetime [yrs]']  - table_lifetime['H2 drydep lifetime [yrs]'] 
table_lifetime
```

```python
table_to_manuscript = table_lifetime.drop('CNTR')
table_to_manuscript.index = [sub.replace('antro', 'anthro') for sub in table_to_manuscript.index]
table_to_manuscript.index = [sub.replace('maxdep', 'lowlatdep') for sub in table_to_manuscript.index]
table_to_manuscript = table_to_manuscript.drop(['delta tot [yrs]','delta atm [yrs]','delta drydep [yrs]'],axis=1)
table_to_manuscript
```
